```python
def outlier(data,feature):
    IQR=data[feature].quantile(0.75)-data[feature].quantile(0.25)
    LB= data[feature].quantile(0.25)-(IQR*1.5)
    UB= data[feature].quantile(0.75)+(IQR*1.5)
    data.loc[data[feature]>=UB,feature]=UB
```

## Handling Null Values of Categorical columns

```python
cat.info()
```

```python
cat=cat.drop(['Alley','PoolQC','Fence','MiscFeature','FireplaceQu'],axis=1)
```

```python
cat.isnull().sum()
```

```python
def impute_cat(data,feature):
    cat[feature+'_random']=cat[feature]
    null= data[feature].isnull().sum()
    random_sample=data[feature].dropna().sample(null)
    random_sample.index=data[data[feature].isnull()].index
    data.loc[data[feature].isnull(),feature+'_random']=random_sample
```

```python
impute_cat(cat,'MSZoning')
impute_cat(cat,'Utilities')
impute_cat(cat,'Exterior1st')
impute_cat(cat,'Exterior2nd')
impute_cat(cat,'MasVnrType')
impute_cat(cat,'BsmtQual')
impute_cat(cat,'BsmtCond')
impute_cat(cat,'BsmtExposure')
impute_cat(cat,'BsmtFinType1')
impute_cat(cat,'BsmtFinType2')
impute_cat(cat,'Electrical')
impute_cat(cat,'KitchenQual')
impute_cat(cat,'Functional')
impute_cat(cat,'GarageType')
impute_cat(cat,'GarageFinish')
impute_cat(cat,'GarageQual')
impute_cat(cat,'GarageCond')
impute_cat(cat,'SaleType')
```
